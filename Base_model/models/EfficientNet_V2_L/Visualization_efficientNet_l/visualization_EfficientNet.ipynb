### 10% training data

```python
import optuna
from optuna.visualization import (
    plot_optimization_history,
    plot_param_importances,
    plot_parallel_coordinate,
)

# Load the study
study = optuna.load_study(study_name="Efficient_V2_optuna_winding_large_ten", storage="sqlite:///Efficient_V2_optuna_winding_large_ten.db")

new_study = optuna.create_study(direction=study.direction)

for trial in study.trials[1:]:
    new_study.add_trial(trial)
    
# Plot optimization history
history_fig = plot_optimization_history(new_study)
history_fig.show()

# Plot parameter importances
importances_fig = plot_param_importances(new_study)
importances_fig.show()

# Plot parallel coordinate
parallel_coordinate_fig = plot_parallel_coordinate(new_study)
parallel_coordinate_fig.show()
```
